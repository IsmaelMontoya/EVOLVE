## 💡 Solución Ejercicio 1: El Contador Global
**Problema:** Python asume que `contador` dentro de la función es una nueva variable local, pero falla al intentar sumarle 1 porque no tiene valor inicial.
**Solución:** Usamos la palabra clave `global` para decirle a Python que queremos modificar la variable que vive fuera de la función.

```python
contador = 0

def actualizar_contador():
    global contador  # <--- Esta es la clave
    contador = contador + 1

actualizar_contador()
print(f"Contador final: {contador}")
```

## 💡 Solución Ejercicio 3: El Validador (StaticMethod)
**Clave:** No usamos `__init__` ni `self` porque son funciones puras que no guardan estado.

```python
class Validator:
    
    @staticmethod
    def is_valid_pass(password):
        # Devuelve True si la longitud es mayor a 8
        return len(password) > 8
    
    @staticmethod
    def is_adult(age):
        return age >= 18

# Pruebas sin instanciar
print("Pass '123' válida?", Validator.is_valid_pass("123"))
print("Pass 'supersegura123' válida?", Validator.is_valid_pass("supersegura123"))
print("Es adulto con 20?", Validator.is_adult(20))
```

## 💡 Solución Ejercicio 4: La Caja Fuerte (Privacidad Básica)
**Clave:** Usar el doble guion bajo `__` para ocultar la contraseña.

```python
class Safe:
    def __init__(self, password, money):
        self.money = money          # Público (Cualquiera lo ve)
        self.__password = password  # Privado (Oculto)

caja = Safe("1234", 100000)

print(f"Dinero en la caja: {caja.money}€") # Funciona

# Intentamos acceder a la contraseña
try:
    print(caja.__password)
except AttributeError as e:
    print(f"❌ ERROR DE ACCESO: {e}")
```

## 💡 Solución Ejercicio 5: Login Seguro (Properties)
**Clave:** El Getter enmascara la salida y el Setter valida la entrada.

```python
class User:
    def __init__(self, username, password):
        self.username = username
        self.__password = password # Guardamos la real en privado

    # GETTER: Cuando pidan .password, les damos asteriscos
    @property
    def password(self):
        return "*******"

    # SETTER: Cuando intenten cambiarla, validamos
    @password.setter
    def password(self, new_pass):
        if len(new_pass) > 6:
            print("✅ Contraseña actualizada correctamente.")
            self.__password = new_pass
        else:
            print("❌ Error: La contraseña es demasiado corta.")

u = User("admin", "secret")

print(f"Leyendo contraseña: {u.password}")

print("Intentando cambiar a '123'...")
u.password = "123"

print("Intentando cambiar a 'nuevasecreta'...")
u.password = "nuevasecreta"
```

## 💡 Solución PROYECTO FINAL: Sistema Modular
Generamos el archivo y lo usamos integrando todo lo aprendido.

```python
%%writefile empleados.py

class Employee:
    def __init__(self, name, salary):
        self._name = name
        self.__salary = salary

    @property
    def salary(self):
        return self.__salary

    @salary.setter
    def salary(self, amount):
        if amount < 0:
            print("❌ Error: El salario no puede ser negativo.")
        else:
            print(f"✅ Salario actualizado a {amount}")
            self.__salary = amount

    @staticmethod
    def calculate_tax(amount):
        return amount * 0.15
```

```python
# 1. Importar
import empleados as emp

# 2. Crear objeto
empleado = emp.Employee("Carlos", 2000)

# 3. Intento fallido de salario negativo
print("--- Test Salario Negativo ---")
empleado.salary = -500

# 4. Cambio correcto
print("\n--- Test Salario Correcto ---")
empleado.salary = 3000

# 5. Uso de método estático (sin usar el objeto 'empleado', sino la clase)
print("\n--- Cálculo de Impuestos ---")
impuestos = emp.Employee.calculate_tax(3000)
print(f"Para un salario de 3000, los impuestos son: {impuestos}")
```
