## 2. Definition and Syntax

Una función es un bloque de código reutilizable. Observa cómo ahora indicamos qué tipo de dato entra y qué tipo sale.

**Sintaxis con Type Hints:**
```python
def function_name(param: type) -> return_type:
    # Code block
    return result
```

````python
# Basic Example
def greet() -> None:
    print("Hello world from a function!")

# Calling the function
greet()
````

### 2.1 Parameters and Arguments
- **Parameter:** La variable definida en el paréntesis.
- **Argument:** El valor real enviado.

````python
def greet_person(name: str) -> None:
    print(f"Hello, {name}, welcome to Python.")

greet_person("Ana")      # 'Ana' is the argument
greet_person("Peter")
````

---

## 3. The `return` statement

Las funciones devuelven valores. Si no hay `return`, devuelven `None`.

````python
def km_to_miles(km_distance: float) -> float:
    miles = km_distance * 0.621371
    return miles

# Storing the result in a variable
distance = km_to_miles(10.0)
print(f"10 km are {distance} miles")
````

### 3.1 Multiple Return
Python devuelve múltiples valores como una tupla.

````python
def calculate_square_and_cube(number: int) -> Tuple[int, int]:
    square = number ** 2
    cube = number ** 3
    return square, cube

res_sq, res_cb = calculate_square_and_cube(3)
print(f"Square: {res_sq}, Cube: {res_cb}")
````

## 4. Argument Types

### 4.1 Positional and Keywords
Podemos pasar argumentos por orden o por nombre.

````python
def describe_person(name: str, age: int, city: str) -> None:
    print(f"{name} is {age} years old and lives in {city}.")

# Positional (Order matters)
describe_person("Lucy", 30, "Madrid")

# Keywords (Order doesn't matter)
describe_person(city="Barcelona", name="Charles", age=25)
````

### 4.2 Default Values
Valores por defecto si el usuario no envía nada.

````python
def make_coffee(sugar: bool = False) -> None:
    if sugar:
        print("Coffee with sugar ready ☕🍬")
    else:
        print("Black coffee ready ☕")

make_coffee()        # Uses default (False)
make_coffee(True)    # Overwrites default
````

## 5. Variable Arguments (`*args` & `**kwargs`)

* **`*args`**: Tupla de argumentos posicionales.
* **`**kwargs`**: Diccionario de argumentos nombrados.

````python
# Example using *args (Tuple)
def sum_all(*numbers: int) -> int:
    print(f"Received: {numbers} (Type: {type(numbers)})")
    return sum(numbers)

print("Total sum:", sum_all(1, 2, 3, 4, 10))
````

````python
# Example using **kwargs (Dictionary)
def user_data(**data: Any) -> None:
    print(f"Data received: {data} (Type: {type(data)})")
    for key, value in data.items():
        print(f"- {key}: {value}")

user_data(name="John", age=33, vip=True)
````

## 6. Scope (Variable Visibility)

* **Local:** Dentro de la función.
* **Global:** Fuera de la función.

````python
x: str = "GLOBAL"  # Global variable

def test_function() -> None:
    x: str = "LOCAL"  # Local variable (shadows global)
    print("Inside:", x)

test_function()
print("Outside:", x)
````

---

## 7. Lambda Functions (Anonymous)

Funciones de una línea sin nombre.

**Sintaxis:** `lambda args: expression`

````python
# Regular function
def double(x: int) -> int:
    return x * 2

# Equivalent Lambda
double_lambda = lambda x: x * 2

print(double(5))
print(double_lambda(5))
````

---

## 8. Recursion

Una función que se llama a sí misma. Requiere un **base case**.

````python
# Classic Example: Factorial
# 5! = 5 * 4 * 3 * 2 * 1

def factorial(n: int) -> int:
    if n == 1:       # Base case
        return 1
    else:
        return n * factorial(n - 1)  # Recursive call

print("Factorial of 5:", factorial(5))
````

## 9. Documentation (Docstrings)
Documentación estándar explicando argumentos y retornos.

````python
def triangle_area(base: float, height: float) -> float:
    """
    Calculates the area of a triangle.
    
    Args:
        base (float): The base of the triangle.
        height (float): The height of the triangle.
        
    Returns:
        float: The calculated area.
    """
    return (base * height) / 2

# Checking help
help(triangle_area)
````
