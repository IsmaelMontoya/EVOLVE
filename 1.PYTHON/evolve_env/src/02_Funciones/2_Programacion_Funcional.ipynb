```python
# Functions can be assigned to variables and passed around

def add_one(x):
    return x + 1

def apply_to_list(fn, items):
    return [fn(x) for x in items]

numbers = [1, 2, 3]
print("Original:", numbers)
print("Add one:", apply_to_list(add_one, numbers))
```

## `map()`, `filter()` y `reduce()`
- `map(fn, iterable)` transforma cada elemento.
- `filter(fn, iterable)` filtra por condición.
- `reduce(fn, iterable)` acumula valores y devuelve uno solo (desde `functools`).


```python
from functools import reduce

numbers = [1, 2, 3, 4, 5]

mapped = list(map(lambda x: x * 2, numbers))
filtered = list(filter(lambda x: x % 2 == 0, numbers))
reduced = reduce(lambda a, b: a + b, numbers)

print("Mapped:", mapped)
print("Filtered:", filtered)
print("Reduced sum:", reduced)
```

## Alternativas prácticas: list comprehensions + any/all
Las comprensiones suelen ser más legibles que `map`/`filter`. `any` y `all` ayudan a evaluar condiciones en colecciones.


```python
# List comprehension vs map/filter
numbers = [1, 2, 3, 4, 5]

squares_map = list(map(lambda x: x ** 2, numbers))
squares_comp = [x ** 2 for x in numbers]

print("Map:", squares_map)
print("Comprehension:", squares_comp)
```

## `sorted()` con `key` y ranking
Podemos ordenar por múltiples criterios usando una tupla en `key`.


```python
# Sorting by average desc, then absences asc

def average(grades):
    return sum(grades) / len(grades)

sorted_students = sorted(
    students,
    key=lambda s: (-average(s["grades"]), s["absences"]),
)

print(sorted_students)
```
