## 1️⃣ Series: La columna vertebral

Una **Serie** es un array unidimensional (como una lista) pero con un índice explícito. Es la pieza fundamental de un DataFrame (un DataFrame es una colección de Series).

Podemos crear una Serie desde una lista o un diccionario.

```python
# 1. Creating from a List
values = [10, 20, 30, 40]
s_list = pd.Series(values, index=['a', 'b', 'c', 'd'])

print("Series from List:\n", s_list)
print("\nAccessing by label 'b':", s_list['b'])
```

```python
# 2. Creating from a Dictionary
# Keys become the Index, Values become the Data
data_dict = {'Apple': 100, 'Banana': 200, 'Cherry': 150}
s_dict = pd.Series(data_dict)

print("Series from Dict:\n", s_dict)
```

### Atributos de una Serie
Es útil separar los valores del índice.

```python
print("Values:", s_dict.values)
print("Index:", s_dict.index)
```

## 2️⃣ DataFrames: La Tabla de Datos

Un **DataFrame** es una estructura tabular bidimensional (filas y columnas). Es lo más parecido a una hoja de cálculo.

Generalmente se crean a partir de un **Diccionario de Listas**, donde:
* Las **Claves** (Keys) son los nombres de columnas.
* Los **Valores** (Values) son listas con los datos.

```python
# Creating a DataFrame manually
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [25, 30, np.nan, 40],
    'City': ['New York', 'Paris', 'London', 'Tokyo'],
    'Salary': [50000, 60000, 70000, 80000]
}

df_manual = pd.DataFrame(data)
display(df_manual)
```

## 4️⃣ Inspección de Datos: El primer paso

Antes de analizar nada, debemos entender qué tenemos delante: dimensiones, tipos de datos, valores nulos...

```python
# Let's use a larger dataset (Manual creation for demo purposes if file is missing)
rows_n, cols = np.random.randint(0, 100, size=(100, 5)), ['A', 'B', 'C', 'D', 'E']

df_large = pd.DataFrame(rows_n, columns=cols)

# 1. First rows
display(df_large.head(3))

# 2. Last rows
display(df_large.tail(3))

# 3. Random sample
display(df_large.sample(3))
```

```python
# 4. Structural Info (Dtypes and Nulls)
df_large.info()
```

```python
# 5. Statistical Summary (Mean, Max, Min...)
df_large.describe()
```

### ⚠️ Concepto Clave: Shape y Shape[0]

La propiedad `.shape` nos devuelve una tupla con las dimensiones del DataFrame: `(filas, columnas)`.

Es vital saber usar `shape[0]` para obtener programáticamente el **número de filas** (muy usado en bucles, validaciones o para calcular porcentajes).

```python
# Getting dimensions
dimensions = df_large.shape
print("Shape (Tuple):", dimensions)

# Accessing specific dimensions
rows = df_large.shape[0]
cols = df_large.shape[1]

print(f"\nThe DataFrame has {rows} rows and {cols} columns.")

# Example usage: Percentage
print(f"Processing row 10 out of {rows}...")
```

## 5️⃣ Selección e Indexación: El Bisturí

Pandas ofrece múltiples formas de seleccionar datos. Es fácil confundirse, así que hay que tener claro:
1.  **`[]` (Corchetes):** Selección de Columnas.
2.  **`.iloc`:** Selección por **Posición** (Índice numérico).
3.  **`.loc`:** Selección por **Etiqueta** (Nombre).

```python
display(df_large["A"].head(3))
```

```python
# A) Column Selection
# Returns a Series
col_a = df_large['A']

# Returns a DataFrame (Double brackets)
cols_subset = df_large[['A', 'C', 'E']]

display(cols_subset.head(3))
```

```python
# B) Position Selection (.iloc)
# Syntax: [rows_range, cols_range]
# Remember: Python slices exclude the last number

# First 5 rows, first 3 columns
display(df_large.iloc[0:5, 0:3])

# Specific rows and columns (e.g., row 10, column 2)
print("Value at [10, 2]:", df_large.iloc[10, 2])

# Specific rows and columns (e.g., row 4, column 2)
print("Value at [4, 2]:", df_large.iloc[4, 2])
```

```python
# C) Label Selection (.loc)
# Syntax: [row_names, col_names]
# Note: .loc INCLUDES the last element in the slice

# Rows 0 to 5 (inclusive), Columns 'A' and 'D'
display(df_large.loc[0:5, ['A', 'D']])
```

## 6️⃣ Filtrado Básico (Máscaras Booleanas)

El poder real de Pandas es filtrar filas basándonos en condiciones (similar al `WHERE` en SQL).

```python
# 1. Create a boolean mask
# This returns a Series of True/False
mask = df_large['A'] > 50
print("Mask sample:\n", mask.head())

# 2. Apply the mask to the DataFrame
filtered_df = df_large[mask]
display(filtered_df.head(3))

print(f"\nOriginal rows: {df_large.shape[0]}")
print(f"Filtered rows: {filtered_df.shape[0]}")
```

```python
# 1. Create a boolean mask
# This returns a Series of True/False
filtered_df = df_large[df_large['A'] > 50]
display(filtered_df.head(3))
print(f"\nOriginal rows: {df_large.shape[0]}")
print(f"Filtered rows: {filtered_df.shape[0]}")
```

```python
# Multiple conditions
# AND (&), OR (|), NOT (~)
# Parentheses are MANDATORY

complex_filter_df = df_large[(df_large['A'] > 50) & (df_large['B'] < 20) & ~(df_large['C'] <= 50)][["A","B"]]
display(complex_filter_df.head())
complex_filter_df.shape[0]
```

## 7️⃣ Ordenación (Sorting)

Podemos ordenar el DataFrame por los valores de una o varias columnas.

```python
# Ascending sort
df_sorted = df_large.sort_values(by='A', ascending=False) # Ascending by default is True
display(df_sorted.head())
```

```python
# Sorting by multiple columns
df_multi_sort = df_large.sort_values(by=['A', 'B'], ascending=[False, True])
display(df_multi_sort.head())
```
