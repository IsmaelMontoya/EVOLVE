```python
import pandas as pd
import numpy as np

# ⚠️ SETUP: Generating dummy data to make solutions runnable
# We recreate 'df_large' with 100 rows and columns A, B, C, D, E
np.random.seed(42) # For reproducibility
df_large = pd.DataFrame(np.random.randint(0, 100, size=(100, 5)), columns=['A', 'B', 'C', 'D', 'E'])

print("Setup complete. df_large shape:", df_large.shape)
```

## 💡 Solución Ejercicio 1: El Cirujano de Datos (Selección)

**Enunciado:**
1. Selecciona las columnas 'B' y 'E'.
2. Muestra las filas de la 50 a la 55 usando `.iloc`.
3. Muestra el valor de la fila 0 y la columna 'A' usando `.loc`.

```python
# 1. Select Columns 'B' and 'E'
# We use double brackets for multiple columns
cols_subset = df_large[['B', 'E']]
display(cols_subset.head())
```

```python
# 2. Rows 50 to 55 using .iloc
# Remember: Python slices exclude the stop value, so we go up to 56
rows_subset = df_large.iloc[50:56, :]
display(rows_subset)
```

```python
# 3. Value at Row 0, Column 'A' using .loc
# .loc uses LABELS (names)
value = df_large.loc[0, 'A']
print("Value at [0, 'A']: ", value)
```

---

## 💡 Solución Práctica Final: Manipulación Completa

**Enunciado:**
1. Crea una nueva columna llamada `Suma` que sea la suma de 'A' + 'B'.
2. Filtra las filas donde `Suma` sea mayor a 100.
3. Ordena el resultado por la columna `C` de forma descendente.
4. Guarda el resultado en un archivo `resultado.csv`.

```python
# 1. Create 'Suma' column
df_large['Suma'] = df_large['A'] + df_large['B']

# 2. Filter rows where 'Suma' > 100
# We create a boolean mask first
mask = df_large['Suma'] > 100
filtered_df = df_large[mask]

print(f"Rows after filtering: {filtered_df.shape[0]} (Original: {df_large.shape[0]}) ")
```

```python
# 3. Sort by 'C' descending
final_df = filtered_df.sort_values(by='C', ascending=False)

display(final_df.head())
```

```python
# 4. Save to CSV
# index=False prevents pandas from saving the row numbers (0, 1, 2...) as a column
final_df.to_csv('resultado.csv', index=False)
print("File 'resultado.csv' saved successfully.")
```
