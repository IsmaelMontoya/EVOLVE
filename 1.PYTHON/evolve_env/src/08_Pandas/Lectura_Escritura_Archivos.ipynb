## 📑 2. Archivos CSV

Un archivo CSV (Comma Separated Values) es una tabla donde cada línea representa una fila, y los valores están separados por comas.

Python incluye el módulo `csv` para trabajar con este tipo de archivos sin instalar nada.

---


````python
# 📂 MANUAL CSV READING (The Hard Way)

file_path = 'personas.csv'
parsed_data = []

# 1. Open the file safely using 'with'
with open(file_path, 'r') as file:
    # 2. Read all lines into a list
    lines = file.readlines()
    
    # 3. Extract and clean the header (first line)
    # .strip() removes the invisible new line character '\n'
    # .split(',') separates the string by commas
    header = lines[0].strip().split(',')
    
    # 4. Iterate through the rest of the lines (Data)
    for line in lines[1:]:
        # Cleaning and splitting each row
        values = line.strip().split(',')
        
        # 5. Create a dictionary for each row (Optional, but cleaner)
        row_dict = {}
        for i in range(len(header)):
            # We map the header name to the value
            row_dict[header[i]] = values[i]
            
        parsed_data.append(row_dict)

# 6. Display results
print(f"Total rows loaded: {len(parsed_data)}")
print("\n--- All rows ---")
for row in parsed_data:
    print(row)
````

````python
# -------------------------------------------------------------------
# Escribir un archivo CSV
# -------------------------------------------------------------------
import csv

datos = [
    ["Nombre", "Edad", "Ciudad"],
    ["Ana", 25, "Lima"],
    ["Luis", 30, "Bogotá"],
    ["María", 22, "Buenos Aires"]
]

with open("personas.csv", "w", newline="", encoding='utf-8') as f:
    escritor = csv.writer(f)
    escritor.writerows(datos)
````

````python
# -------------------------------------------------------------------
# Leer un archivo CSV
# -------------------------------------------------------------------

with open("personas.csv", "r", encoding='latin-1') as f:
    lector = csv.reader(f)
    for fila in lector:
        print(fila)
````

## 🧩 3. Archivos JSON

Los archivos JSON (`.json`) son muy usados para guardar datos estructurados (como diccionarios o listas de diccionarios).

Python trae un módulo llamado `json` para codificar y decodificar este tipo de archivos.

---


````python
# -------------------------------------------------------------------
# Escribir un archivo JSON
# -------------------------------------------------------------------
import json

persona = {
    "nombre": "Ana",
    "edad": 25,
    "intereses": ["lectura", "programación"]
}

with open("persona.json", "w", encoding='utf-8') as f:
    json.dump(persona, f, indent=4)
````

````python
gente = [
    {"nombre": "Ana", "edad": 25, "intereses": ["lectura", "programación"]},
    {"nombre": "Luis", "edad": 30, "intereses": ["deportes", "música"]},
    {"nombre": "María", "edad": 22, "intereses": ["arte", "cine"]}
]

with open("gente.json", "w", encoding='utf-8') as f:
    json.dump(gente, f, indent=4)  # Escribir la lista de personas en el archivo JSON

    '''el parámetro indent controla la indentación (sangría) del JSON generado,
    es decir, cuántos espacios se usan para anidar niveles.'''
````

````python
# -------------------------------------------------------------------
# Leer un archivo JSON
# -------------------------------------------------------------------

with open("persona.json", "r") as f:
    datos = json.load(f)

for k,v in datos.items():
    print(f"{k.title()}: {v}")
````

````python
with open("gente.json", "r") as f:
    datos = json.load(f)
for persona in datos:
    for k,v in persona.items():
        print(f"{k.title()}: {v}")
    print("---")
````

````python
with open("gente.json", "r") as f:
    datos = json.load(f)
print(datos)
print(datos[0])
print(datos[0]['intereses'])
print(datos[0]['intereses'][0])

print('-----------------------------------------------------')
for persona in datos:
    print("Nombre:", persona["nombre"])
    print("Edad:", persona["edad"])
    print("Intereses:", persona["intereses"])
    print()
````
